```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the linear function and its derivative
def linear_function(x):
    return x

def linear_derivative(x):
    return np.ones_like(x)

# Define the nonlinear function and its derivative
def nonlinear_function(x):
    return x**2

def nonlinear_derivative(x):
    return 2 * x
```

```python
# Generate a range of x values
x = np.linspace(-10, 10, 400)

# Compute the function values and their derivatives
y_linear = linear_function(x)
y_linear_derivative = linear_derivative(x)
y_nonlinear = nonlinear_function(x)
y_nonlinear_derivative = nonlinear_derivative(x)
```

```python
# Plot the linear function and its derivative
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y_linear, label='Linear: f(x) = x')
plt.plot(x, y_linear_derivative, label="Linear Derivative: f'(x) = 1")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Linear Function and Its Derivative')
plt.legend()

# Plot the nonlinear function and its derivative
plt.subplot(1, 2, 2)
plt.plot(x, y_nonlinear, label='Nonlinear: f(x) = x^2')
plt.plot(x, y_nonlinear_derivative, label="Nonlinear Derivative: f'(x) = 2x")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Nonlinear Function and Its Derivative')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

```python
x0=0
y_linear_tangent=linear_function(x0)+linear_derivative(x0)*x

x0=0
y_nonlinear_tangent=nonlinear_function(x0)+nonlinear_derivative(x0)*x
```

```python
# Plot the linear function and its tangent

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y_linear, label='Linear: f(x) = x')
plt.plot(x, y_linear_tangent, label="Linear Derivative: f'(x) = 1")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Linear Function and Its Tangent')
plt.legend()

# Plot the nonlinear function and its tangent

plt.subplot(1, 2, 2)
plt.plot(x, y_nonlinear, label='Nonlinear: f(x) = x^2')
plt.plot(x, y_nonlinear_tangent, label="Nonlinear Derivative: f'(x) = 2x")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Nonlinear Function and Its Tangent')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

#2D

```python
def function(x,y):
  return x**2 - y**2 + 2

def Dfunction_x(x,y):
  return 2*x

def Dfunction_y(x,y):
  return - y*2
```

```python
x = np.linspace(-20, 20, 100)
y = np.linspace(-20, 20, 100)

X,Y=np.meshgrid(x,y)
Z=function(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='binary')

plt.title('2D function')

# Display the plots
plt.show()
```

```python
def T(X,Y,x0,y0):
  return function(x0,y0)+Dfunction_x(x0,y0)*X+Dfunction_y(x0,y0)*Y
```

```python
x0=0
y0=1

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z)

ax.plot_surface(X,Y,T(X,Y,x0,y0))

plt.title('2D function')

# Display the plots
plt.show()
```
